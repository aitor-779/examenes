# Ejercicio 4

```python
ana=alumnos['analisis']
dis=alumnos['diseno']
imp=alumnos['implementacion']
pre=alumnos['presentacion']
coe=alumnos['coevaluacion']
media=(ana+dis+imp+pre+coe)/5
alumnos['NotaFinal']= media
filtro1=alumnos['analisis']>=5
filtro2=alumnos['diseno']>=5
filtro3=alumnos['implementacion']>=5
filtro4=alumnos['presentacion']>=5
filtro5=alumnos['coevaluacion']>=5
filtro=filtro1&filtro2&filtro3&filtro4&filtro5
print(alumnos.loc[(filtro),['nombre','apellidos','NotaFinal']])
```
